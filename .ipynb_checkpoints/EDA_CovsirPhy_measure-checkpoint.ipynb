# S-R Trend Analysis with CovsirPhy

```python
# imports 
import covsirphy as cs
import matplotlib.pyplot as plt
```

```python
# loading the data -> stored in kaggle/input
# data will be downloaded if existing files are older than 24h
data_loader = cs.DataLoader(directory="kaggle/input")

# The number of cases and population values in jhu format
jhu_data = data_loader.jhu()

# Government Responses
oxcgrt_data = data_loader.oxcgrt()
```

```python
#cases and deaths whole dataset
total_df = jhu_data.total()

total_df.tail()
```

```python
#gov response whole dataset
oxcgrt_data.layer().tail()
```

```python
#take switzerland subset of gov response dataframe
gov_switzerland = oxcgrt_data.subset("Switzerland")
gov_switzerland.tail()
```

```python
#esample visualization of measure of interest
#Visualization of stringency_index
oxcgrt_data.map(variable="Stringency_index")
```

```python
#visualize measure over time and correlation with #cases #deaths

#choose neighboring countries as benchmark
neigh = ["Switzerland","Germany","Italy","France","Austria"]
cond = oxcgrt_data.layer()["Country"].isin(neigh)
neighbors_df = oxcgrt_data.layer()[cond]
```

```python
#plots
## create figure and axes
fig, (ax1,ax2) = plt.subplots(2,1)
fig, countries_ax = plt.subplots(len(neigh),1)
ax1.scatter(data=neighbors_df,
          x="Date",
          y="Stringency_index",
          c=neighbors_df.Country.astype("category").cat.codes)
ax1.legend()

#choose best visualization
for i,country in enumerate(neigh):
    #subset measure by country
    country_measures = neighbors_df[neighbors_df.Country==country]
    #measure - date
    ax2.plot(neighbors_df[neighbors_df.Country==country].Date,
             country_measures.Stringency_index,label=country)
    #subset cases by country
    country_cases, complement = jhu_data.records(country)
    #measure - cases
    countries_ax[i].scatter(country_cases.Date,
                country_cases.Infected,
                c=country_measures[country_measures.Date.isin(country_cases.Date)].Stringency_index.astype("category").cat.codes)
    countries_ax[i].set_title(country)
ax2.legend(loc=(1.04,0))
```
