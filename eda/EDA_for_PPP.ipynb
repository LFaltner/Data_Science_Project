# Plots for PowerPointPresentation

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datasets = ['cases_covid_data', 'face_covering_policies_covid', 'stay_at_home_covid' ]

cases = pd.read_csv('data/cases_covid_data.csv')
lockd = pd.read_csv('data/stay_at_home_covid.csv')
masks = pd.read_csv('data/face_covering_policies_covid.csv')
```

```python
# slice SUI

cases_sui = cases[cases['location'] == 'Switzerland'].copy()
lockd_sui = lockd[lockd['Entity'] == 'Switzerland'].copy()
masks_sui = masks[masks['Entity'] == 'Switzerland'].copy()

# set to pd datetime

cases_sui['date'] = pd.to_datetime(cases_sui.iloc[:,3])
lockd_sui['Day'] = pd.to_datetime(lockd_sui.iloc[:,2])
masks_sui['Day'] = pd.to_datetime(masks_sui.iloc[:,2])

# set new index

cases_sui = cases_sui.set_index(cases_sui['date'])
lockd_sui= lockd_sui.set_index(lockd_sui['Day'])
masks_sui = masks_sui.set_index(masks_sui['Day'])

# slice data to same length

lockd_sui = lockd_sui[lockd_sui.index >= min(cases_sui.index)]
masks_sui = masks_sui[masks_sui.index >= min(cases_sui.index)]
cases_sui = cases_sui[cases_sui.index <= max(lockd_sui.index)]



```

```python
display(cases_sui)
display(lockd_sui)
display(masks_sui)
```

```python
# plot data

cases_avg = cases_sui['new_cases_smoothed'].copy()
lockd = lockd_sui['stay_home_requirements'].copy()
masks = masks_sui['facial_coverings'].copy()
```

```python
fig = plt.figure(figsize = (15,5))
plt.plot(cases_avg.index, cases_avg, label = 'Smoothed Cases', color = 'black')
plt.scatter(cases_avg.index, cases_avg, c = lockd, alpha = 0.5, s = 15, cmap = 'viridis')
plt.xlabel('Date')
plt.ylabel('Smoothed Cases')
cbar = plt.colorbar(ticks = [0,1])
cbar.set_label(' No Lockdown / Lockdown')
plt.title('Lockdown state over time')
plt.savefig('Lockdown_EDA')
```

```python
fig = plt.figure(figsize = (15,5))
plt.plot(cases_avg.index, cases_avg, label = 'Smoothed Cases', color = 'black')
plt.scatter(cases_avg.index, cases_avg, c = masks, alpha = 0.5, s = 15)
cbar = plt.colorbar(ticks = [0,1,2,3])
plt.xlabel('Date')
plt.ylabel('Smoothed Cases')
cbar.set_label('Mask state')
plt.title('Mask state over time')
plt.savefig('Mask_EDA')
```
