```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
cases = pd.read_csv('wrangled_data/cases_CH.csv')
lockd = pd.read_csv('wrangled_data/lockdown_CH.csv')
masks = pd.read_csv('wrangled_data/masks_CH.csv')

lockd['stay_home_requirements'] = pd.factorize(lockd['stay_home_requirements'])[0]
masks['facial_coverings'] = pd.factorize(masks['facial_coverings'])[0]

masks.head()
lockd.head()
cases.head()
```

```python
plt.scatter(lockd['Day'], lockd['stay_home_requirements'] * 100000, color= 'green')
plt.plot(cases['date'], cases['new_cases'])
plt.plot(masks['Day'], masks['facial_coverings']*10000)
```
